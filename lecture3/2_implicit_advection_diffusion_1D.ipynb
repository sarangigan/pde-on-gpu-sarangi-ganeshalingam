```julia
#Task 1
using Plots, Plots.Measures, Printf
default(size=(1200, 800), framestyle=:box, label=false, grid=false, margin=10mm, lw=6, labelfontsize=20, tickfontsize=20, titlefontsize=24)

@views function implicit_advection_diffusion_1D()
    # physics
    lx      = 20.0
    dc      = 1.0
    da      = 1000
    vx      = 1.0
    # numerics
    nx      = 100
    tol    = 1e-8
    maxiter = 50nx
    ncheck  = ceil(Int, 0.25nx)
    nt      = 10
    # derived numerics
    dx      = lx / nx
    dt      = dx/abs(vx)
    da      = lx^2/dc/dt
    re      =  + sqrt(^2 + da)
           = (lx / (dc * re))^2
    
    xc      = LinRange(dx / 2, lx - dx / 2, nx)
    d      = dx / sqrt(1 / )
    # array initialisation
    C       = @. 1.0 + exp(-(xc - lx / 4)^2) - xc / lx
    C_i     = copy(C)
    qx      = zeros(Float64, nx - 1)
    C_old   = copy(C)
    # iteration loop
    iter = 1; err = 2tol; iter_evo = Float64[]; err_evo = Float64[]
    anim2 = @animate for it = 1:nt
        C_old .= C
        iter   = 1; err = 2tol; iter_evo = Float64[]; err_evo = []
        
        while err >= tol && iter <= maxiter
            
            qx         .-= d ./ ( .+ d / dc) .* (qx ./ dc .+ diff(C) ./ dx)
            C[2:end-1] .-= d ./ (1.0 .+ d / dt) .* ((C[2:end-1] .- C_old[2:end-1]) ./ dt .+ diff(qx) ./ dx)
            
            if iter % ncheck == 0
                err = maximum(abs.(diff(dc .* diff(C) ./ dx) ./ dx .- (C[2:end-1] .- C_old[2:end-1]) ./ dt))
                push!(iter_evo, iter / nx); push!(err_evo, err)

            end

        iter += 1
            
        end
        C[2:end]   .-=  dt  .* diff(C) ./ dx .* vx
            p1 = plot(xc, [C_i, C]; xlims=(0, lx), ylims=(-0.1, 2.0),
                      xlabel="lx", ylabel="Concentration", title="iter/nx=$(round(iter/nx,sigdigits=3))")
            p2 = plot(iter_evo, err_evo; xlabel="iter/nx", ylabel="err",
                      yscale=:log10, grid=true, markershape=:circle, markersize=10)
            display(plot(p1, p2; layout=(2, 1)))
    end
    savefig("plot2.png")
    gif(anim2,"anim2.gif";fps=2)
end

implicit_advection_diffusion_1D()
```
